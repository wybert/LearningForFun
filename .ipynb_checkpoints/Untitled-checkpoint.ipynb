```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

## 此行代码设置绘图style
plt.style.use('ggplot')

## 下面是常规的绘图命令
df1 = pd.read_csv('beautifulPlot/dataforexcies.csv')
df1 = df1[['a','b']][:5]
df1.plot(kind='bar',title='bar plot');
```

```python
tips_data = pd.read_csv('tips.csv')
```

```python
tips_data.head()
```

```python
import seaborn as sns
g = sns.FacetGrid(tips_data, row="sex", col="smoker")
g.map(plt.hist, "total_bill");
```

```python
g = sns.FacetGrid(tips_data, row="sex", col="smoker")
g.map(sns.kdeplot, "total_bill");
```

```python
g = sns.FacetGrid(tips_data, row="sex", col="smoker")
g.map(plt.scatter, "total_bill", "tip");
```

```python
g = sns.FacetGrid(tips_data, row="sex", col="smoker", margin_titles=True)
g.map(sns.regplot, "total_bill", "tip", order=2);
```

```python
g = sns.FacetGrid(tips_data, row="sex", col="smoker")
g.map(plt.scatter, "total_bill", "tip")
g.map(sns.kdeplot, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips_data, row="sex", col="smoker", hue="day")
g.map(plt.scatter, "tip", "total_bill")
g.add_legend()
```

```python
from matplotlib import cm
fig, ax = plt.subplots()
cax = plt.imshow(df1.values,interpolation='nearest' ,cmap=cm.coolwarm);
cbar = fig.colorbar(cax, ticks=[ 0.05279744, 0.58088577, 0.86400646])
cbar.ax.set_yticklabels(['< -1', '0', '> 1'])  # vertically oriented colorbar
```
